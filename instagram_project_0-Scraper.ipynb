## Scraping list of instagram profiles links

```python
# Creating a function for scraping links on the influencers' instagram accounts
def influencers_inst(n_pages,countries_lst,
                     url='https://starngage.com/app/us/influencer/ranking/',on=False):
    
    """
    Scrape data from https://starngage.com/
    Takes:
        url='https://starngage.com/app/us/influencer/ranking/' - (str)
        n_pages - (int) - number of page to scrape
        countries_lst - (list of str) - list of countries users from
        on=False - (bool) - on(True) or off(False) function
        
    """
    if on:
        result = []
        for country in tqdm(countries_lst):
            for n in tqdm(range(1,n_pages+1)):
                current_url = url + country +'?page='+ str(n)
                res = requests.get(current_url)
                if res.status_code != 200:
                    print(f'Error in access {res.status_code}')
                    break
                soup = BeautifulSoup(res.text,'lxml')
                table = soup.find('table',{'class':"table table-hover table-responsive-sm"})
                for row in table.find('tbody').find_all('tr'):
                    person = {}
                    person['name'] = row.find('td',{'class':"align-middle text-break"}).text.split('@')[0]
                    person['user_id'] = 'https://www.instagram.com/'+row.find('a',{'target':"_blank"}).text[1:]+'/'
                    person['country'] = row.find('a',{'class':"link"}).text
                    person['categories'] = [category.text for category in row.find_all('span',
                                            {'class':"badge badge-pill badge-light samll text-muted"})]
                    person['link'] = row.find('a').attrs['href']
                    result.append(person)
        pd.DataFrame(result).to_csv('./datasets/full_posts.csv',index=False)
        return pd.DataFrame(result)
            
```

#### Dividing list of links by 100 and save to txt

```python
# Splitting links by 100 per file
# Saving as .txt files for scraping it with OctoParse
def n_links_txt(df,n,on = False):
    '''
    Takes:
    df - DataFrame
    n - (int) - number of rows in one txt file
    on=False - (bool) - on(True) or off(False) function 
    
    Saves rows from df in len(df)/n files by n rows
    '''
    if on:
        start = 0
        stop = n
        for i in tqdm(range(math.ceil(len(df)/n))):
            df[['user_id']][start:stop].to_csv('./data/inst{}.txt'.format(i), 
                                               header=None, index=None, sep=' ', mode='a')
            start +=n
            stop +=n
```

## Uploading scraped files to one main file

#### Open each csv file and add the information from the file to the main_df_posts

```python
def post_comments(n_files,on=False):
    """
    Open each csv file and add the information from the file to the main df
    
    Takes:
    n_files (int) number of files to open
    on=False (bool) - on(True) or off(False) function 
    
    Returns:
    main_df_posts - new merged df with all posts from influencers
    """
    if on:
        main_df_posts = pd.DataFrame()
        for i in range(n_files):

            post_path = './datasets/posts_info/'+str(i)+'.csv'
            df_post = pd.read_csv(post_path,error_bad_lines=False,usecols=['Title', 
            'Titlesing', 'alt1', 'Locsing', 'likesing', 'Datesing',
           'Viewsing', 'timesoing', 'profile', 'posts', 'name', 'description',
           'link', 'Page_URL1', 'comment1', 'comment2', 'comment3', 'comment4',
           'comment5', 'Text6', 'comment7', 'comment9', 'Text9', 'comment11',
           'Text11', 'comment12', 'count_comm', 'alt3', 'alt2', 'carousel'])
            main_df_posts = main_df_posts.append(df_post)
        main_df_posts.to_csv('./datasets/main_df_posts.csv',index=False)
    
        return main_df_posts
```

```python
%%time
# Using function for merging all separate files
main_df_posts = post_comments(40,on=False)
```

#### Read json files with info about posts from instagram

```python
def comments_df(on=False):
    """
    Read json files with info about posts from instagram
    Extracting:
    - num_likes - number of likes
    - num_comments - number of comments
    - post_type - type of post: GraphSidecar=carousel, GraphVideo=video, GraphImage=image
    - photo_id - id of post
    
    Takes:
    on=False (bool) - on(True) or off(False) function 
    
    Returns:
    df with info about posts
    """
    if on:
        comments = []
        for path in glob.glob('./datasets/comments_number/*.json'):
            file_name = path.split("\\")[1]
            with open('./datasets/comments_number/'+file_name,encoding='utf8') as f:
                data = json.load(f)
                for i in data['GraphImages']:
                    one_file = {}
                    one_file['num_likes'] = i['edge_media_preview_like']['count']
                    one_file['num_comments'] = i['edge_media_to_comment']['count']
                    one_file['post_type'] = i['__typename']
                    one_file['user_name'] = i['username']
                    one_file['photo_id'] = i['shortcode']
                    one_file['timestamp'] = i['taken_at_timestamp']
                    
                    try:
                        one_file['caption_text'] = i['edge_media_to_caption']['edges'][0]['node']['text']
                    except:
                        one_file['caption_text'] = None
                    comments.append(one_file)
        comments = pd.DataFrame.from_dict(comments)
        comments.to_csv('./datasets/main_df_comments.csv',index=False)
        return comments
```

## Scrape most popular hashtags

#### Download 10000 most popular hashtags

```python
# Creating a function for scraping hashtags and number of posts used these hashtags
def hashtags_populatisy(n_pages,on=False):
    
    """
    Scrape hashtags and number of posts used these hashtags
    from https://top-hashtags.com/instagram/
    Takes:
        n_pages - (int) - number of page to scrape (one page contain 100 hashtags)
        on=False - (bool) - on(True) or off(False) function
        
    """
    if on:
        result = []
        current_url = 'https://top-hashtags.com/instagram/'
        for _ in tqdm(range(n_pages)):
            res = requests.get(current_url)
            if res.status_code != 200:
                print(f'Error in access {res.status_code}')
            soup = BeautifulSoup(res.text,'lxml')
            table = soup.find('ul',{'class':"i-group"})
            for row in table.find_all('li',{'class':"i-row"}):
                hashtags = {}
                hashtags['hashtag'] = row.find('div',{'class':"i-tag"}).find('a').attrs['href'].split('/')[-2]
                if row.find('div',{'class':"i-total"}).text[-1] == 'B':
                    hashtags['popularity'] = int(float(row.find('div',{'class':"i-total"}).text[:-1])*1_000_000_000)
                elif row.find('div',{'class':"i-total"}).text[-1] == 'M':
                    hashtags['popularity'] = int(float(row.find('div',{'class':"i-total"}).text[:-1])*1_000_000)
                else:
                    hashtags['popularity'] = row.find('div',{'class':"i-total"}).text
                result.append(hashtags)
            table = soup.find('div',{'class':"i-paging"})
            tag = table.find('div',{'class':"i-next"}).find('a').attrs['href']
            current_url = 'https://top-hashtags.com/instagram/' + tag.split('/')[-2]
            
        pd.DataFrame(result).to_csv('./datasets/hasgtags_popularity.csv',index=False)
        return pd.DataFrame(result)
            
```

#### Download 30 similar hashtags 

```python
# downloading 30 similar hashtags for each popular main hashtag
result = []
for hashtag in tqdm(hashtags['hashtag'].values):
    current_url = 'http://best-hashtags.com/hashtag/' + hashtag
    res = requests.get(current_url)
    if res.status_code != 200:
        print(hashtag)
        print(f'Error in access {res.status_code}')
    soup = BeautifulSoup(res.text,'lxml')
    table = soup.find('tbody')
    hashtags_similar = {'main_hashtag':hashtag,
                           'similar':[],
                           'num_posts':[]}
    try:
        for row in table.find_all('tr'):
            hashtags_similar["similar"].append(row.find('a').attrs['href'].split('/')[-2])
            num_posts = str(row.find_all('td')[-1])
            clean_num_posts = num_posts.replace('<td>','').replace('</td>','').replace(',','')
            hashtags_similar["num_posts"].append(clean_num_posts)
        result.append(hashtags_similar)
    except:
        continue
#     sleep(randint(1,2))
```
